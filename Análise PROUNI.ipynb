## Tratamento da base

```python
# Apagar as colunas
df.drop(columns=['CODIGO_CURSO', 'INSCRICAO_ENEM', 'CPF', 'MATRICULA', 'CODIGO_CAMPUS', 'CODIGO_IES', 'OPCAO', 'PERCENTUAL_BONUS', 'TP_COTA'], inplace = True)
```

```python
# Substituir a vírgula para o ponto
df = df.apply(lambda x: x.str.replace(',', '.') if x.dtype == "object" else x)
```

```python
# Alterar o tipo de dados das colunas
nota_cols = df.filter(regex='^NOTA').columns
peso_cols = df.filter(regex='^PESO').columns
media_cols = df.filter(regex='^MEDIA').columns


df[nota_cols] = df[nota_cols].astype(float)
df[peso_cols] = df[peso_cols].astype(float)
df[media_cols] = df[media_cols].astype(float)
```

```python
# Preencher os valores nulos pela média
for column in df.select_dtypes(include=['float', 'int']).columns:
    mean_value = df[column].mean()  # Calcula a média da coluna
    df[column].fillna(mean_value, inplace=True)
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt

colunas_notas = ['NOTA_L', 'NOTA_CH', 'NOTA_CN', 'NOTA_M', 'NOTA_R',
                 'NOTA_L_COM_PESO', 'NOTA_CH_COM_PESO', 'NOTA_CN_COM_PESO',
                 'NOTA_M_COM_PESO', 'NOTA_R_COM_PESO']

# Criação dos histogramas
plt.figure(figsize=(20, 10))

for i, coluna in enumerate(colunas_notas, 1):
    plt.subplot(2, 5, i)  # 2 linhas, 5 colunas de gráficos
    plt.hist(df[coluna], bins=10, alpha=0.7, color='skyblue', edgecolor='black')
    plt.title(f'Distribuição de {coluna}')
    plt.xlabel('Nota')
    plt.ylabel('Frequência')

plt.tight_layout()
plt.show()
```

```python
# Contagem de aprovações por sexo
contagem_aprovados = df.groupby(['SEXO', 'APROVADO']).size().unstack(fill_value=0)

# Calculando os percentuais de aprovação por sexo
percentuais_aprovados = (contagem_aprovados['S'] / contagem_aprovados.sum(axis=1)) * 100

# Gráfico de barras
plt.figure(figsize=(8, 6))
percentuais_aprovados.plot(kind='bar', color=['pink', 'skyblue'])
plt.title('Percentual de Aprovação por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Percentual de Aprovação (%)')
plt.xticks(rotation=0)
plt.show()
```

```python
# Convertendo os dados de tipo de concorrência para categoria (para melhor visualização)
df['TIPO_MOD_CONCORRENCIA'] = df['TIPO_MOD_CONCORRENCIA'].astype('category')

# Criando o gráfico boxplot
plt.figure(figsize=(12, 6))

# Plotando o boxplot para as diferentes notas mínimas exigidas
sns.boxplot(data=df[['NOTA_MINIMA_L', 'NOTA_MINIMA_CH', 'NOTA_MINIMA_CN', 'NOTA_MINIMA_M', 'NOTA_MINIMA_R']],
            palette='Set2',
            orient='h')

# Adicionando título e rótulos
plt.title('Comparação das Notas Mínimas Exigidas por Tipo de Concorrência', fontsize=14)
plt.xlabel('Nota Mínima Exigida', fontsize=12)
plt.ylabel('Tipos de Concorrência', fontsize=12)

# Exibindo o gráfico
plt.show()
```

```python
# Contagem de aprovados por curso
aprovados_por_curso = df[df['APROVADO'] == 'S'].groupby('NOME_CURSO').size()

# Selecionando os 10 cursos com maior número de aprovados
top_10_cursos_aprovados = aprovados_por_curso.sort_values(ascending=True).tail(10)

# Criando o gráfico de barras horizontal
top_10_cursos_aprovados.plot(kind='barh', figsize=(12, 6), color='lightgreen')

# Adicionando título e rótulos
plt.title('Top 10 Cursos com Mais Aprovados', fontsize=14)
plt.xlabel('Quantidade de Aprovados', fontsize=12)
plt.ylabel('Curso', fontsize=12)

# Exibindo o gráfico
plt.show()
```

```python
# Contagem de não aprovados por curso
aprovados_por_curso = df[df['APROVADO'] == 'N'].groupby('NOME_CURSO').size()

# Selecionando os 10 cursos com maior número de não aprovados
top_10_cursos_aprovados = aprovados_por_curso.sort_values(ascending=True).tail(10)

# Criando o gráfico de barras horizontal
top_10_cursos_aprovados.plot(kind='barh', figsize=(12, 6), color='lightcoral')

# Adicionando título e rótulos
plt.title('Top 10 Cursos com Mais Reprovados', fontsize=14)
plt.xlabel('Quantidade de Reprovados', fontsize=12)
plt.ylabel('Curso', fontsize=12)

# Exibindo o gráfico
plt.show()
```

```python
# Contagem de aprovados por UF
aprovados_por_uf = df[df['APROVADO'] == 'S'].groupby('UF_CANDIDATO').size()

# Contagem de total de candidatos por UF
total_por_uf = df.groupby('UF_CANDIDATO').size()

# Calculando o índice de aprovação (proporção de aprovados)
indice_aprovacao = (aprovados_por_uf / total_por_uf)*100

# Selecionando as UFs com maior índice de aprovação
top_uf_aprovados = indice_aprovacao.sort_values(ascending=True).tail(10)

# Criando o gráfico de barras horizontal
fig, ax = plt.subplots(figsize=(12, 6))  # Definindo a variável ax corretamente
top_uf_aprovados.plot(kind='barh', color='lightblue', ax=ax)

# Adicionando título e rótulos
plt.title('Top 10 UFs com Maior Índice de Aprovados', fontsize=14)
plt.xlabel('Índice de Aprovação', fontsize=12)
plt.ylabel('UF', fontsize=12)

# Adicionando os rótulos de dados
for index, value in enumerate(top_uf_aprovados):
    ax.text(value + 0.02, index, f'{value:.2f}', va='center', fontsize=10)

# Exibindo o gráfico
plt.show()
```

```python
# Agrupando os cursos e pegando a maior nota de corte para cada um
notas_de_corte_por_curso = df.groupby('NOME_CURSO')['NOTA_CORTE'].max()

# Selecionando os 10 cursos com as maiores notas de corte
top_cursos_nota_corte = notas_de_corte_por_curso.sort_values(ascending=True).tail(10)

# Criando o gráfico de barras horizontal
fig, ax = plt.subplots(figsize=(12, 6))  # Definindo o tamanho do gráfico
top_cursos_nota_corte.plot(kind='barh', color='lightgreen', ax=ax)

# Adicionando título e rótulos
plt.title('Top 10 Cursos com Maiores Notas de Corte', fontsize=14)
plt.xlabel('Nota de Corte', fontsize=12)
plt.ylabel('Curso', fontsize=12)

# Adicionando os rótulos de dados
for index, value in enumerate(top_cursos_nota_corte):
    ax.text(value + 2, index, f'{value:.2f}', va='center', fontsize=9)

# Exibindo o gráfico
plt.show()
```
